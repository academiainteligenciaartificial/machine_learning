## Operaciones con arreglos (array)

```python
np.random.seed(0)
array1 = np.random.randint(10, size=(2, 2))
array1, array1.shape
```

```python
np.random.seed(10)
array2 = np.random.randint(10, size=(2, 2))
array2, array2.shape
```

```python
array3 = array1 + array2
array3
```

```python
np.random.seed(5)
array4 = np.random.randint(10, size=(2, 2, 3))
array4
```

```python
array5 = array2 * array4
array5
```

## Reshape

```python
array2.shape, array4.shape
```

```python
array2_reshape = array2.reshape(2, 2, 1)
array2_reshape
```

```python
array2_reshape.shape
```

```python
array5 = array2_reshape * array4
array5
```

## Transpose

```python
np.random.seed(8)
array6 = np.random.randint(10, size=(2, 5))
array6
```

```python
array7 = array6.T
array7
```
