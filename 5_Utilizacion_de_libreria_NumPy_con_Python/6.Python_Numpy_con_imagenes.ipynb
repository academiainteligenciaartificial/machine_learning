```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_carro = plt.imread("car_c.jpg")
```

```python
array_carro.shape
```

```python
plt.imshow(array_carro, interpolation='nearest')
```

```python
capa_rojo = array_carro[:, :, 0]
plt.imshow(capa_rojo, cmap=plt.cm.Reds_r)
```

```python
capa_verde = array_carro[:, :, 1]
plt.imshow(capa_verde, cmap=plt.cm.Greens_r)
```

```python
capa_verde.shape
```

```python
capa_azul = array_carro[:, :, 1]
plt.imshow(capa_azul, cmap=plt.cm.Blues_r)
```
