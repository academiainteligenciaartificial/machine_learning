# Convertir a minúsculas

Primero convertimos los valos a string con la función ***str***

```python
titanic_dataset["Name"] = titanic_dataset["Name"].str.lower()
```

```python
titanic_dataset.head()
```

# Completar los valores NaN

Para completar los valores NaN utilizaremos la función ***fillna***

```python
titanic_dataset["Cabin"] = titanic_dataset["Cabin"].fillna("C000")
```

```python
titanic_dataset.head()
```

# Agregando una nueva columna a nuestro DataFrame

```python
titanic_dataset = pd.read_csv("titanic.csv")
titanic_dataset.head()
```

```python
nacionalidades = pd.Series(["US", "UK", "FR"]) #Primero creamos nuestra Series
titanic_dataset["Nationality"] = nacionalidades
titanic_dataset.head()
```

Podemos asignar un valor único a todas las filas

```python
titanic_dataset["Number of baggages"] = 2
titanic_dataset.head()
```

# Eliminar una columna

Para eliminar una columna utilizamos la función ***drop()***, si nos referimos a una columna utilizamos el parametro ***axis=1***

```python
titanic_dataset.drop("Number of baggages", axis=1, inplace=True)
titanic_dataset.head()
```

# Obteniendo muestras de nuestro dataset

Con la función ***sample*** podemos obtener una muestra según una fracción enviada con el atributo ***frac=***. Esta función primero baraja (shuffle) el data set y obtiene la muestra, si enviamos el valor 1 seria el mismo dataset

```python
titanic_dataset_shuffled = titanic_dataset.sample(frac=1)
titanic_dataset_shuffled
```

Para regrear a las posiciones originales utilizaremos la función ***reset_index()***

```python
# Por defecto la función crea una nueva columna con los index anteriores
# el atributo drop=True eliminara la columna creada de los index anteriores
titanic_dataset_shuffled.reset_index(inplace=True, drop=True)
titanic_dataset_shuffled
```

# Funciones lambda

Las funciones lambda son funciones anónimas (sin nombre) y son definidas con la keyword ***lambda***, estas funciones estan compuestas por tres partes.
1. Parametros de entrada
2. Dos puntos
3. Proceso definido en una línea\
EJM.\
lambda x: x + 1

```python
# Multiplicaremos el precio del ticket por 10 utilizando la función apply 
```

```python
titanic_dataset.head()
```

```python
titanic_dataset["Fare"] = titanic_dataset["Fare"].apply(lambda x: x * 10)
titanic_dataset
```
