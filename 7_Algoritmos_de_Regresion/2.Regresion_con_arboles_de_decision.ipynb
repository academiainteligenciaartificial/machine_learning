# Dataset

```python
import pandas as pd
weather = pd.read_csv('lima_weather_full.csv')
weather
```

```python
weather_filtered = weather[['datetime', 'temp']]
weather_filtered
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(weather_filtered['datetime'].index, weather_filtered['temp'], alpha=0.5)
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

```python
weather_filtered[45:55]
```

```python
weather_filtered = weather_filtered.drop([50, 51])
weather_filtered
```

```python
plt.figure()
plt.scatter(weather_filtered['datetime'].index, weather_filtered['temp'], alpha=0.5)
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

```python
X = weather_filtered['datetime'].index
y = weather_filtered['temp']
print(len(X), len(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
import numpy as np
X_train = np.reshape(X_train, (-1, 1))
print(X_train.shape)
```

```python
dtr_model = DecisionTreeRegressor()
dtr_model.fit(X_train, y_train)
```

```python
y_pred = dtr_model.predict(X_test[:, np.newaxis])
```

```python
values = np.column_stack([X_test, y_pred])
df_values = pd.DataFrame(values, columns=['X_test', 'y_pred'])
df_values.sort_values(by=['X_test'], inplace=True)
df_values
```

```python
plt.figure()
plt.scatter(X_test, y_test, alpha=0.5, color='green')
plt.scatter(X_test, y_pred, alpha=0.5, color='red')
plt.plot(df_values['X_test'], df_values['y_pred'], alpha=0.5, color='red')
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

# Evaluacion del modelo

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
print(mse)
```
