# Algoritmo de regresi√≥n linear

```python
from sklearn import datasets
```

```python
X, y = datasets.make_regression(
    n_samples=100,
    n_features=1,
    noise=5,
    random_state=42
)
```

```python
import matplotlib.pyplot as plt
```

```python
_ ,ax = plt.subplots()
ax.scatter(X, y, alpha=0.5)
ax.set_xlabel("Caracteristicas / Features")
ax.set_ylabel("Variable objetivo / Target")
```

```python
import numpy as np
```

```python
X = np.interp(X, (X.min(), X.max()), (0, 5))
```

```python
y = np.interp(y, (y.min(), y.max()), (78000, 150000))
```

```python
_ ,ax = plt.subplots()
ax.scatter(X, y, alpha=0.5)
ax.set_xlabel("A√±os de experiencia")
ax.set_ylabel("Salario anual en $")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
_ ,ax = plt.subplots()
ax.plot(X_train, y_train_pred, color='green', alpha=0.5)
ax.scatter(X, y, alpha=0.5)
ax.set_xlabel("A√±os de experiencia")
ax.set_ylabel("Salario anual en $")
```

## Evaluar nuestro algoritmo

```python
y_test_pred = model.predict(X_test)
```

```python
_ ,ax = plt.subplots()
ax.scatter(X_test, y_test, alpha=0.5)
ax.scatter(X_test, y_test_pred, alpha=0.5)
ax.set_xlabel("A√±os de experiencia")
ax.set_ylabel("Salario anual en $")
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_true=y_test, y_pred=y_test_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
